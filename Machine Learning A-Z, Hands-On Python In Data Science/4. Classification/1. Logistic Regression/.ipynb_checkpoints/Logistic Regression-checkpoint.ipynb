### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib widget
%matplotlib notebook
```

### Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1:].values
```

### Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0,)
```

### Feature Scaling
#### Not necessarily required here, but how ever it will imporve the performance of the model.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

### Training the logistic regression model on training data

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

### Predicting a new Result

```python
y_pred = classifier.predict(x_test)
```

```python
plt.plot([i for i in range(len(x_test))], y_test)
plt.plot([i for i in range(len(x_test))], y_pred)
```
